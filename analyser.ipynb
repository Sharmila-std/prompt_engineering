```python
import re
import pandas as pd

data = pd.read_csv("llama-3.2-1b-gsm8k-fullpreprocessed.csv")

# Regex to check messages that end with: The final answer is: $\boxed{number}$
pattern = r"The final answer is: \$\\boxed\{[\d\.]+\}\$$"

# Find invalid responses
invalid_mask = ~data['message'].astype(str).str.strip().str.contains(pattern)
invalid_count = invalid_mask.sum()
invalid_entries = data[invalid_mask]

print(f"Total records: {len(data)}")
print(f"Records NOT ending with the expected pattern: {invalid_count}")
```

```python
import re
import pandas as pd

data = pd.read_csv("llama-3.2-3b-gsm8k_with_fewshot-fullpreprocessed.csv")

# Regex to match and extract the number inside \boxed{...}
pattern = r"The final answer is: \$\\boxed\{([\d\.]+)\}\$$"

# Extract the number if the pattern matches, else NaN
data["final answer"] = data['message'].astype(str).str.strip().str.extract(pattern)

# Count how many messages didn't match the expected pattern
invalid_mask = data["final answer"].isna()
invalid_count = invalid_mask.sum()

print(f"Total records: {len(data)}")
print(f"Records NOT ending with the expected pattern: {invalid_count}")
```

```python
# Ensure both columns are float
data["ground_truth"] = pd.to_numeric(data["ground_truth"], errors='coerce')
data["final answer"] = pd.to_numeric(data["final answer"], errors='coerce')

# Compare and filter matching rows
matching = data[data["ground_truth"] == data["final answer"]]

print(f"Correctly matched records: {len(matching)}")
```
