```python
import pandas as pd
import ast

file_path = "llama-3.2-1b-gsm8k-full"
data = pd.read_csv(file_path+".csv")
print("Lenght of the data",len(data))
ground_truth = pd.read_csv("ground-truth.csv")

data['response_dict'] = data['response'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

data['id'] = data['response_dict'].apply(lambda x: x.get('id'))
data['object'] = data['response_dict'].apply(lambda x: x.get('object'))
data['created'] = data['response_dict'].apply(lambda x: x.get('created'))
data['model'] = data['response_dict'].apply(lambda x: x.get('model'))
data['system_fingerprint'] = data['response_dict'].apply(lambda x: x.get('system_fingerprint'))

data['finish_reason'] = data['response_dict'].apply(lambda x: x.get('choices', [{}])[0].get('finish_reason'))
data['role'] = data['response_dict'].apply(lambda x: x.get('choices', [{}])[0].get('message', {}).get('role'))
data['prompt_tokens'] = data['response_dict'].apply(lambda x: x.get('usage', {}).get('prompt_tokens'))
data['completion_tokens'] = data['response_dict'].apply(lambda x: x.get('usage', {}).get('completion_tokens'))
data['total_tokens'] = data['response_dict'].apply(lambda x: x.get('usage', {}).get('total_tokens'))
data['message'] = data['response_dict'].apply(lambda x: x.get('choices', [{}])[0].get('message', {}).get('content'))

Copymerged_data = pd.merge(
    data, 
    ground_truth,
    left_on="question",   
    right_on="question", 
    how="inner"
)

print("Lenght of the data",len(Copymerged_data))
Copymerged_data.to_csv(file_path+"preprocessed.csv", index=False)
```
